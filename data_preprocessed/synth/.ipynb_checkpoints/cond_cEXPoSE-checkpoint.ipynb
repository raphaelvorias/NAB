```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.decomposition import PCA
from scipy.linalg import subspace_angles
from numpy import linalg as LA
from scipy import stats
from sklearn.kernel_approximation import RBFSampler
import time

def mm(a,b):
    return np.matmul(a,b)
def inv(a):
    return np.linalg.inv(a)

class EXPoSE_VRLS(object): 
    def __init__(self, n):
        self.G = 0.001 * np.eye(n)
        self.weights = np.zeros((n,n))
        self.lambd = 0.99
        
    def update(self,x,y):
        J = np.asscalar(inv(self.lambd+mm(x,mm(self.G,x.T))))
        K = mm(self.G,x.T)
        L = mm(x,self.G)
        self.G = 1/self.lambd*self.G - 1/self.lambd*J*mm(K,L)
        
        M = mm(x,self.weights)-y
        self.weights = self.weights - mm(self.G,mm(x.T,M))
        
    def getCovar(self):
        return self.weights
    
class EXPoSE_VRLS2(object): 
    def __init__(self, n):
        self.G = 0.001 * np.eye(n)
        self.weights = np.zeros((n,n))
        self.lambd = 0.99
        
    def update(self,x,y):
        x = x.T
        alpha = y - mm(x.T,self.weights)
        g = mm(mm(self.G,x),inv(self.lambd+mm(mm(x.T,self.G),x)))
        self.G = 1.0/self.lambd*self.G-mm(mm(g,x.T),1.0/self.lambd*self.G)
        self.weights = self.weights + alpha*g
        
    def getCovar(self):
        return self.weights
    
class EXPoSE_VRLS3(object): 
    def __init__(self, n):
        self.G = 0.001 * np.eye(n)
        self.weights = np.zeros((n,n))
        self.lambd = 0.99
        
    def update(self,x,y):
        x = x.T
        alpha = y - mm(x.T,self.weights)
        g = mm(mm(self.G,x),inv(self.lambd+mm(mm(x.T,self.G),x)))
        self.G = 1.0/self.lambd*self.G-mm(mm(g,x.T),1.0/self.lambd*self.G)
        self.weights = self.weights + alpha*g
        
    def getCovar(self):
        return self.weights
    
class VRLS4(object): 
    def __init__(self, n):
        self.P = 0.001 * np.eye(n)
        self.W = np.zeros((n,n))
        self.lammy = 0.99
        self.iLammy = 1/self.lammy
        
    def update(self,U,D):
        pi = mm(self.P,U)
        K = pi/(self.lammy+mm(U.T,pi))
        epsi = D - mm(self.W.T,U)
        self.W = self.W + mm(K,epsi.T)
        self.P = self.iLammy*self.P - self.iLammy*mm(K,mm(U.T,self.P))
        
    def getCovar(self):
        return self.W
```

```python
df = pd.read_csv('baselineMixedMixed.csv')
df = df.drop('timestamp',axis=1)
df = (df-df.min(axis=0))/(df.max(axis=0)-df.min(axis=0))
#df = (df-df.mean(axis=0))/(df.std(axis=0))
df.head()
```

```python
ansco = []
fourierFeatures = 100

k = RBFSampler(gamma=0.5, n_components=fourierFeatures, random_state=5)
r = EXPoSE_VRLS3(fourierFeatures)

for i in range(1,len(df)):
    x_t = k.fit_transform(df[i-1:i].values)
    y_t = k.fit_transform(df[i:i+1].values)
    
    conditional_mean = (np.matmul(x_t,r.getCovar()))

    anomalyScore = np.clip(abs(np.asscalar(1 - np.inner(y_t, conditional_mean))), 0.0, 1.0)
    ansco.append(anomalyScore)
    
#     if i < 1000:
#         r.update(x_t,y_t)
#     elif anomalyScore<0.9:
#         r.update(x_t,y_t)
        
    r.update(x_t,y_t)

plt.plot(ansco)
```

```python
start = time.time()

ansco = []
fourierFeatures = 500

k = RBFSampler(gamma=0.5, n_components=fourierFeatures, random_state=5)
r = EXPoSE_VRLS3(fourierFeatures)

for i in range(1,len(df)):
    x_t = k.fit_transform(df[i-1:i].values)
    y_t = k.fit_transform(df[i:i+1].values)
    
    conditional_mean = np.matmul(x_t,r.getCovar())
    if i > 1:
        conditional_mean = conditional_mean/LA.norm(conditional_mean)

    anomalyScore = np.asscalar(1 - np.inner(y_t, conditional_mean))
    ansco.append(anomalyScore)
    
    r.update(x_t,y_t)
    
end = time.time()
print(end - start)

plt.plot(ansco)
```
