```python
#https://stackoverflow.com/a/13105359/6028830
def byteify(input):
    if isinstance(input, dict):
        return {byteify(key): byteify(value)
                for key, value in input.iteritems()}
    elif isinstance(input, list):
        return [byteify(element) for element in input]
    elif isinstance(input, unicode):
        return input.encode('utf-8')
    else:
        return input
```

```python
#set up paths and folders
root = os.path.abspath('..')
rawDir = os.path.join(root,'data_raw')
dataDir = os.path.join(root,'data_preprocessed')
labelDir = os.path.join(root,'labels')
print("root     : " + root)
print("raws     : " + rawDir)
print("dataDir  : " + dataDir)
print("labelDir : " + labelDir)
```

```python
#construct a dict per collection of datasets
#the idea is that each collection has datasets with similar formats
collections = {
    "nabArtNo" : {'folder':'/nab/artificialNoAnomaly/','delimiter':',','timestamp':True},
    "nabArtYes" : {'folder':'/nab/artificialWithAnomaly/','delimiter':',','timestamp':True},
    #"nabRealAd" : {'folder':'/nab/realAdExchange/','delimiter':',','timestamp':True},
    #"nabAWSCloud" : {'folder':'/nab/realAWSCloadwatch/','delimiter':',','timestamp':True},
    "nabRealCause" : {'folder':'/nab/realKnownCause/','delimiter':',','timestamp':True},
    #"nabRealTraffic" : {'folder':'/nab/realTraffic/','delimiter':',','timestamp':True},
    #"nabRealTweets" : {'folder':'/nab/realTweets','delimiter':',','timestamp':True},
    #"millAE" : {'folder':'/mill/AE/','delimiter':';','timestamp':True},
    #"millForces" : {'folder':'/mill/Forces/','delimiter':'\t','timestamp':True},
    #"FE" : {'folder':'/FE/','delimiter':',','timestamp':True},
    #"nabsub" : {'folder':'/nab_subselection/','delimiter':',','timestamp':True},
    "mv" : {'folder':'/mv/','delimiter':',','timestamp':False}
}

#now use the collections to create preprocessors and automatically preprocess each folder
for c in collections:
    rawPaths = absoluteFilePaths(rawDir+collections[c]['folder'])
    for f in rawPaths:
        print('\n **Preprocessing ' + f)
        pre = Preprocessor(fileName=f,
                           delimiter=collections[c]['delimiter'],
                           timestamp=collections[c]['timestamp'],
                           folderStruct=collections[c]['folder'])
        pre.autoPP()
print('\n **Done')
```

```python
#create a dict via nab.corpus that contain filenames as keys and dataframes as values
corpus = Corpus(dataDir)
data = corpus.dataFiles
for d in data:
    print(d)
print(data)

#load labels
with open('/home/raph/ad-thesis/labels/combined_labels.json') as json_file:  
    labels = byteify(json.load(json_file))

#set detectors
#detectors = {'null':NullDetector}
#detectors = {'eg_skyline':EarthgeckoSkylineDetector}
detectors = {'expose':ExposeDetector,'eg_skyline':EarthgeckoSkylineDetector,'numenta':NumentaDetector}
#detectors = {'numenta':NumentaDetector}
```

```python
print('\n ==Running detectors on datasets ============================')
plotFileName = "%06x" % random.randint(0, 0xFFFFFF)
runner = Runner(dataSets=data, detectors=detectors, plotFileName=plotFileName)
runner.run()
```
